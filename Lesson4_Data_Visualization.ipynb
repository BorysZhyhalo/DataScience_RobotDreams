### Basic formatting and styles

#### Line colors
More abouts colors is [here](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
plt.plot(x, np.sin(x - 0), color='blue')        # by name
plt.plot(x, np.sin(x - 1), color='g')           # by short name (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # shades of grey from 0 to 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB format, each from 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # CSS color names
```

#### Line style

```python
plt.plot(x, x + 0, linestyle='solid', linewidth=2)
plt.plot(x, x + 1, linestyle='dashed', linewidth=3)
plt.plot(x, x + 2, linestyle='dashdot', linewidth=4)
plt.plot(x, x + 3, linestyle='dotted', linewidth=5);
```

```python
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdota
plt.plot(x, x + 7, linestyle=':');  # dotted
```

### Multiple plots on the one figure


#### MatLab Style

```python
plt.figure()  # create a figure

# The first plot
plt.subplot(2, 1, 1) # (nrows, ncolumns, index)
plt.plot(x, np.sin(x))

# The second plot
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.tight_layout();
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x));

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x));
```

This interface is stateful: it keeps track of the "current" figure and axes on which all plt commands are applied.
You can get a reference to them using the procedures plt.gcf() (get the current image) and plt.gca() (get the current axes).

Although this stateful interface is fast and convenient for simple graphs, it is easy to run into problems with it.
For example, after creating the second panel, how can we go back and add something to the first one?
This is possible in a MATLAB-style interface, but rather inconvenient. Fortunately, there is a second way.

#### Object-Oriented Interface

An object-oriented interface is useful for more complex situations, as well as when you need more control over your figure.
In an object-oriented interface, the plotting functions are methods of the explicit Figure and Axes objects, and do not depend on any notion of an "active" figure or axes.

```python
# ax is array of two Axes objects
fig, ax = plt.subplots(2)

# Виклик plot() метода для потрібного обʼєкта
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
fig, ax = plt.subplots(ncols = 2)

# Виклик plot() метода для потрібного обʼєкта
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

### Scatterplot

a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
# using plt.plot
plt.plot(x, np.sin(x), 'o', color='green');
```

There are lots of markers and additional settings!

```python
plt.figure(figsize=(8,5))
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5)*2, rng.rand(5)*2, marker,
             label=f"marker={marker}", linewidth=4)
plt.legend(loc = 'upper right')
plt.xlim(0, 1.8);
```

```python
plt.plot(x, np.sin(x), '-ob', markersize=3, linewidth=1);
```

Alternatively we can use plt.scatter:

```python
plt.scatter(x, np.sin(x), marker='o');
```

The plot function will be faster for scatterplots where markers don't vary in size or color. But in plt.scatter properties of each individual point (size, fill color, border color, etc.) can be individually controlled or compared to data.

```python
import sklearn.datasets as ds
from sklearn.datasets import load_iris
iris = load_iris()
iris.keys()
```

```python
print(iris['DESCR'])
```

```python
iris.feature_names
```

```python
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar();
```

We can see that this scatterplot gave us the opportunity to explore four different dimensions of the data at the same time:

*   the position (x, y) of each point corresponds to the length and width of the sepal,
*   dot size is related to petal width,
*   and the color depends on the specific type of flower.

### Histogram and density plot

Стовпчики зазвичай використовуються в гістограмі, висота яких по осі Y буде залежати від частоти даних. Тим часом на осі X ми можемо спостерігати досліджувану змінну, розбиту на інтервали (bins).

```python
np.random.seed(10)
data = np.random.randn(1000)
plt.hist(data);
```

Custom histogram as a density plot.
Documentation is [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
plt.hist(data, bins=30, density=True, alpha=0.5, color='steelblue',
         edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

data = [x1, x2, x3]

fig, ax = plt.subplots(3)

ax[0].hist(data, alpha=0.3, density=True, bins=20)
ax[1].hist(data, histtype='stepfilled', alpha=0.3, density=True, bins=40)
ax[2].hist(data, histtype='barstacked', alpha=0.3, density=True, bins=40)

plt.show()
```

### Stacked barplot


```python
A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1.,  2.,  1.,  1.])
X = np.arange(4)

plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'g', bottom = A)
plt.bar(X, C, color = 'r', bottom = A + B)

plt.show()
```

```python
men_mean_scores = (20, 35, 30, 35, 27)
women_mean_scores = (25, 32, 34, 20, 25)
ind = np.arange(5) # the x locations for the groups
width = 0.35

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, men_mean_scores, width, color='r')
ax.bar(ind, women_mean_scores, width, bottom=men_mean_scores, color='b')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
# так встановлюємо текстові лейбли
ax.set_xticks(ind) # values
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5']) # labels
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women'])
plt.show()
```

**Box plot** is very useful plot. It visualizes the main characteristics such as median, percentiles, min and max values of features.

```python
categorical_cols =['Gender', 'Vehicle_Damage']
numeric_cols = [ 'Age', 'Vintage']
```

```python
from tqdm import notebook

def plt_box_plot(dataset, columnName):
    """
    Plots boxplots for column given as parameter.
    """
    dataset.boxplot(by="Response", column=columnName)
    plt.tight_layout();

for column in notebook.tqdm(numeric_cols, desc = "Our charts are being ready"):
    plt_box_plot(df, column)
```

```python
df.boxplot(column="Age", by=["Vehicle_Age", "Response"], figsize=(10,8))
plt.tight_layout()
plt.show()
```
